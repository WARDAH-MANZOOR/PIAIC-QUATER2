```python
plt.plot(x,y, 'k--') #Color of the plot, line style
plt.xlabel('x')
plt.ylabel('sinx')
plt.grid(True, which='major') # major matlab wo points jese yahn 20,40 hain 
plt.title('Graph of Sin x')
```

```python
x=np.linspace(1,10,100) # 1 se 10 tak ka data 100 ki divison ke sath
y=x+2
plt.plot(x,y, 'k--') #Color of the plot, line style
plt.xlabel('x')
plt.ylabel('sinx')
plt.grid(b=True, which='major', color='b', linestyle='-')  # major matlab wo points jese yahn 20,40 hain
plt.grid(b=True, which='minor', color='r', linestyle='--') # minor wo lines hoti hain jo points humare pas hain matlab 20,40 tou inke darmiyan lines create karti hain
plt.minorticks_on()
plt.title('Graph of Sin x')
# here red lines are minor lines
# blue lines are major
# inse hum points behtar andaz me mark karsakte
```

```python
#Subplots: Having more than one plots in a figure
```

```python
x = np.linspace(0, 2*np.pi, 400) #400 equidistant points between 0 and 2*pi
y = np.sin(x**2)
f, (ax1, ax2) = plt.subplots(1, 2) #Two figures in one row
ax1.plot(x, y) # normal plot
ax1.set_title('Figure 1') #Setting title of first figure
ax2.scatter(x, y) # scatterplot
ax2.set_title('Figure 2') #Setting title of second figure
```

```python
# HISTOGRAM 
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000) 
data = pd.DataFrame(data, columns=['x', 'y']) 
for col in 'xy': 
    plt.hist(data[col], alpha=0.5)
```

```python
#KERNEL DENSITY PLOT
for col in 'xy': 
    sns.kdeplot(data[col], shade=True)
```

```python
# Distplot
sns.distplot(data['x']) 
sns.distplot(data['y'])
```

```python
# PAIR PLOT
iris = sns.load_dataset("iris") 
iris.head()
sns.pairplot(iris, hue='species', size=2.5);
```

```python
# FACTOR PLOT
tips = sns.load_dataset('tips') 
tips.head()
with sns.axes_style(style='ticks'): 
        g = sns.factorplot("day", "total_bill", "sex", data=tips, kind="box") 
        g.set_axis_labels("Day", "Total Bill");
```
