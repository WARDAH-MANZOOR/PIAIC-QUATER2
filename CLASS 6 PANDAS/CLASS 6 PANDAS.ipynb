```python
import numpy as np
mean = frame['Total (100)'].mean() #MEAN
std = frame['Total (100)'].std() # STANDARD DEVIATION 
lower_range = mean - 3*std
upper_range = mean + 3*std
```

```python
filtered_data = frame[frame['Total (100)'] < upper_range][frame['Total (100)'] > lower_range]
```
