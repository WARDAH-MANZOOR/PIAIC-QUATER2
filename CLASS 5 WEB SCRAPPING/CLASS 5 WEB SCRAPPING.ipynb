```python
# printing first ranking


#Print ranking for only one format
for th in trs[0].find_all('th'): #Heading row (first row)
        print('{:13}'.format(th.text), end=" ") #====> 13 is characer width mean the gap of two words
print() #Move to next line
for tr in trs[1:]:  #All rows exclusding first row because all rows have td but first row have th so we do separately for first row the above code is for first row  
    for td in tr.find_all('td'): #All columns
        print('{:13}'.format(td.text), end=" ")
    print()
```

```python
#Now print rankings for all the format
```

```python
def print_ranking(trs):
    for th in trs[0].find_all('th'): #Heading row
        print('{:20}'.format(th.text), end=" ")
    print() #Move to next line
    for tr in trs[1:]:  #All rows    
        for td in tr.find_all('td'): #All columns
            print('{:20}'.format(td.text), end=" ")
        print()
```

```python
for r in ranking: #For all the rankings
    print(r.caption.text)
    print_ranking(r.find_all('tr'))
    print('\n\n\n')
```

```python
#Let's do the same thing with enumerate
for i,r in enumerate(ranking): #For all the rankings
    print(r.caption.text)
    print(format[i].text)
    print_ranking(r.find_all('tr'))
    print('\n\n\n')
```

```python
def csv_ranking(trs):
    data = ""
    for th in trs[0].find_all('th'): #Heading row
        data+=th.text+","
    data = data[:-1] #Remove last comma
    data+='\n' #Move to next line
    for tr in trs[1:]:  #All rows    
        for td in tr.find_all('td'): #All columns
            data+=td.text+","
        data = data[:-1]
        data+='\n'
    return data
```

```python
r=ranking[0]
data = csv_ranking(r.find_all('tr'))
print(data)
```

```python
r=ranking[0]
data = csv_ranking(r.find_all('tr'))
file = format[0].text # format[0] is file name which is ICC Test Rankings
data = csv_ranking(r.find_all('tr'))
with open(file+".csv","w") as f:
    f.writelines(data)
```

```python
# files for all formats
for i,r in enumerate(ranking):
    file = format[i].text
    data = csv_ranking(r.find_all('tr'))
    with open(file+".csv","w") as f:
        f.writelines(data)
```
