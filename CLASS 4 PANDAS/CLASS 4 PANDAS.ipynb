```python
def calculate_loss_tie(row):
    return row['played'] - row['win']
```

```python
stats['loss_tie']=stats.apply(calculate_loss_tie,axis=1)
#The function calculate_loss will be called four time with following rows:
#1	5	10	Pakistn	5 - row
#t2	6	8	India	4 - row
#t3	4	8	Srilanka	4
#t4	5	6	S. Africa	3
```
